```python
mang_chua_27_tong_diem = []
for i in range(27):
    tong_mon = 0.0
    
    for j in range(soluong):
        tong_mon += df3.iloc[j,i]
        
    mang_chua_27_tong_diem.append(tong_mon/soluong)
tenMon=df3.columns
mang_chua_27_tong_diem
```

```python
a=np.array(mang_chua_27_tong_diem)
a.sum()/27
```

```python
min_27 = min(mang_chua_27_tong_diem)
max_27 = max(mang_chua_27_tong_diem)
hieu = max_27 - min_27


moc_tb = np.linspace(min_27,max_27, 6)
arr_tb =[]
for i in range(len(moc_tb)):
    arr_tb.append(moc_tb[i])


print(arr_tb)



arr_dtbk = []
for j in range(1,len(arr_tb)):
    
    dtbk = (arr_tb[j-1]+ arr_tb[j])/2
    arr_dtbk.append(dtbk)
    
arr_dtbk

```

```python
chia_deu_5_muc_kho = hieu/5
```

```python
kho_5 = min_27 + 1*chia_deu_5_muc_kho
kho_4 = min_27 + 2*chia_deu_5_muc_kho
kho_3 = min_27 + 3*chia_deu_5_muc_kho
kho_2 = min_27 + 4*chia_deu_5_muc_kho
kho_1 = min_27 + 5*chia_deu_5_muc_kho
```

```python
kho_3
```

```python
min_27
```

```python
def do_kho(diem, min_27, kho_5, kho_4, kho_3, kho_2, kho_1):
    if (diem>=min_27 and diem<=kho_5):
        return 1
    elif (diem>kho_5 and diem<=kho_4):
        return 2
    elif (diem>kho_4 and diem<=kho_3):
        return 3
    elif (diem>kho_3 and diem<=kho_2):
        return 4
    else:
        return 5
```

```python
mang_do_kho_mon = {}
for i in range(27):
    x = mang_chua_27_tong_diem[i]
    s = do_kho(x, min_27, kho_5, kho_4, kho_3, kho_2, kho_1)
    mang_do_kho_mon[tenMon[i]]=s
```

```python
mang_do_kho_mon
```

```python
import math
```

```python
def diem_tt(diem,tenMon):
    print(arr_dtbk[mang_do_kho_mon[tenMon]-1])
    print((diem+arr_dtbk[mang_do_kho_mon[tenMon]-1])/mang_do_kho_mon[tenMon])
    return ((diem+arr_dtbk[mang_do_kho_mon[tenMon]-1])/mang_do_kho_mon[tenMon]) + (abs(1-mang_do_kho_mon[tenMon]))/(2)
```
